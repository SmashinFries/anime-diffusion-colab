### Load Waifu Diffusion Model

```python
from diffusers import StableDiffusionPipeline
import torch
from torch import autocast
model_id = "hakurei/waifu-diffusion"
device = "cuda"


pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16, revision='fp16')
pipe = pipe.to(device)

def dummy(images, **kwargs):
  return images, False

# NSFW filter? We aint weak
pipe.safety_checker = dummy
```

### Load Prompt Generator

```python
from transformers import pipeline

gpt2_pipe = pipeline('text-generation', model='Gustavosta/MagicPrompt-Stable-Diffusion', tokenizer='gpt2')
```

# Start UI
Click the url once everything loads

```python
import gradio as gr, random, re
import torch
from torch import autocast
from transformers import pipeline, set_seed




block = gr.Blocks(css=".container { margin: auto; }")

def infer(prompt, num_samples, guidance_scale, steps, height, width):
      with autocast("cuda"):
        images = pipe([prompt] * num_samples, guidance_scale=guidance_scale, num_inference_steps=steps, height=height, width=width)
  
      return images["sample"]

def genPrompt(starting_text, amount):
    for count in range(amount):
        seed = random.randint(100, 1000000)
        set_seed(seed)

        if len(starting_text) < 1:
          return "Provide context first!"
    
        response = gpt2_pipe(starting_text, max_length=random.randint(60, 90), num_return_sequences=amount)
        response_list = []
        for x in response:
            resp = x['generated_text'].strip()
            if resp != starting_text and len(resp) > (len(starting_text) + 4) and resp.endswith((":", "-", "—")) is False:
                response_list.append(resp + "\n")
    
        response_end = "\n".join(response_list)
        response_end = re.sub('[^ ]+\.[^ ]+','', response_end)
        response_end = response_end.replace("<", "").replace(">", "")

        if response_end != "":
            return response_end
        if count == 4:
            return response_end

def clearPromptGen(txt):
    return ""

with block as demo:
    gr.Markdown("<h1><center>KuzuTech Diffusion</center></h1>")
    # gr.Markdown(
    #     "just go for it"
    # )
    with gr.Row():
        with gr.Tab("Generate"):
            # with gr.Box():
                with gr.Row().style(mobile_collapse=False):

                    text = gr.Textbox(
                        label="Enter your prompt", show_label=True, interactive=True,
                    ).style(
                        border=(True, False, True, True),
                        rounded=True,
                        container=True,
                    )
                
                with gr.Row().style(mobile_collapse=False, equal_height=True):
                    btn = gr.Button("Run").style(
                        margin=(5,5,5,5),
                        rounded=(True, True, True, True),
                    )
                

        with gr.Tab("Prompt Helper"):
            with gr.Box():
                with gr.Column():
                        starter_prompt = gr.Textbox(label="Starter Prompt", placeholder="Begin Prompt Here...", interactive=True)
                        prmpt_amount = gr.Slider(1, 15, step=1, value=4, label="Amount of Prompts")
                        btn_gen = gr.Button("Generate Prompt").style(rounded=(True, True, True, True), full_width=True, margin=(5,5,5,5))
                        prompt_gen_clr = gr.Button("Clear Output").style(rounded=True, full_width=True)
                        choices = gr.Textbox(label="Output", interactive=False)

        with gr.Tab("Settings"):

            with gr.Box():
                
                with gr.Row().style(mobile_collapse=False, equal_height=True):
                    # img_batch = gr.Slider(1, 50, step=1, value=1, label="Number of Batches")
                    img_amount = gr.Slider(1, 20, step=1, value=2, label="Number of Images")
                    
                with gr.Row().style():
                    img_steps = gr.Slider(0, 100, step=1, value=20, label="Steps")
                    guidance = gr.Slider(0, 50, step=.5, value=7.5, label="Guidance Scale")

                with gr.Row().style(mobile_collapse=False, equal_height=True):
                    img_height = gr.Slider(0, 3200, step=64, value=512, label="Height")

                    img_width = gr.Slider(0, 3200, step=64, value=512, label="Width")

        gallery = gr.Gallery(label="Generated images", show_label=False).style(grid=[2], height="auto")

               
        
    text.submit(infer, inputs=[text, img_amount, guidance, img_steps, img_height, img_width], outputs=gallery)
    btn.click(infer, inputs=[text, img_amount, guidance, img_steps, img_height, img_width], outputs=gallery)
    prompt_gen_clr.click(clearPromptGen, inputs=[choices], outputs=choices)
    btn_gen.click(genPrompt, inputs=[starter_prompt, prmpt_amount], outputs=choices)

    gr.Markdown(
        """___
   <p style='text-align: center'>
   Model fine-tuned by <a href="https://huggingface.co/hakurei">hakurei</a>
   <br/>
   Colab created by <a href="https://github.com/smashinfries">Smashinfries</a>
   </p>"""
    )


demo.launch(debug=True)
```
